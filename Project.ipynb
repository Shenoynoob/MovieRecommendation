## Data Preprocessing 


```python
movies.isnull().sum()
```

```python
movies.dropna(inplace = True )  #before we had 3 movies with no overview so we dropped them 
```

```python
movies = movies.drop_duplicates()
```

```python
# now changing the format of genres
movies.iloc[0].genres
```

```python
# our desired output is : ['Action','Adventure','fantasy','Scifi']
import ast
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L

movies['genres'] = movies['genres'].apply(convert)  


# error showing beacuse it has already been excecuted once 
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
# now keywords and genres have been formatted successfull 
```

```python
# formatting the cast : including only the top 3 actors
def convert3(obj):
    L = []
    counter = 0 
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1 
        else:
                break
        
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
# now for crew we only need the name of the director 
def fetchDirector(obj):
    L = []
    counter = 0 
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break 
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetchDirector)
```

```python
movies.head()
```

```python
# now since the overview column in a string type and we want to concatenate multiple columns 
# we will first convert the overview column to string 
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# now we will the blank spaces between the words from each columns so that our model dosent get confused between the different
# people having the same first name 
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" " , "")for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" " , "")for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" " , "")for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" " , "")for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview']+ movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['id' , 'title' , 'tags']]
```

```python
new_df.head()
```

```python
new_df['tags']= new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df.head()
```

```python
# now conveting the tags column to lower 
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem(text):
    y =[]

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df.head()
```

## Text Vectorization

```python
# basically we will convert the movies into vectors 
# such that we will have 5000 vectore and if i want a suggestion from a movie , i will get the vectors closest to the movie vector 
# converting text into words using a text vectorization method 
# we are going use BOW(Bag of Words)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000 , stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
# now we have successfully created the vector now we will find the distance between the vectors
# we will use cosine distance instead of eucleadian distance because in large datasets and datasets with very high dimensions eucleadin distance fails 
```

```python
from sklearn.metrics.pairwise import cosine_similarity 
similarity = cosine_similarity(vectors)
```

```python
# function to recommend movies 
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse = True , key = lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
        
  
```

```python
recommend('Avatar')
```
